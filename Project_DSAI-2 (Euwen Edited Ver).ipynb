```python
# Basic Libraries
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt # we only need pyplot
sb.set() # set the default Seaborn style for graphics
```

```python
planet_data = pd.read_csv("planetary_data.csv")
cleaned_data = planet_data.dropna()
print("First few rows of the cleaned data:")
print(cleaned_data.head(10))
```

```python
numDF = pd.DataFrame(cleaned_data[["pl_orbper","pl_orbsmax","pl_rade","pl_bmasse","pl_orbeccen","st_rad","st_mass"]])
numDF.head(50)
```

```python
numDF.describe()
```

```python
numDF.corr()
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, StackingRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Assuming numDF is available and contains 'pl_orbsmax' (semi-major axis) and 'pl_orbper' (orbital period)
X = numDF.drop(columns=["pl_orbper"])  # Semi-major axis as a feature
y = numDF['pl_orbper']     # Target variable: Orbital period

# Transform semi-major axis to a^(3/2) for linear regression with Kepler's law transformation
X_transformed = X ** (3 / 2)

# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X_transformed, y, test_size=0.2, random_state=42)

# Polynomial transformation for polynomial regression model
poly_transformer = PolynomialFeatures(degree=2)
X_train_poly = poly_transformer.fit_transform(X_train)
X_test_poly = poly_transformer.transform(X_test)

# Initialize base models
linear_model = LinearRegression()
poly_model = LinearRegression()  # Used with polynomial features
tree_model = DecisionTreeRegressor(random_state=42)
forest_model = RandomForestRegressor(n_estimators=100, random_state=42)
boosting_model = GradientBoostingRegressor(n_estimators=100, random_state=42)

# Create stacking ensemble with a meta-model
# Using RandomForest as meta-model for stacking
stacking_ensemble = StackingRegressor(
    estimators=[
        ('linear', linear_model),
        ('poly', poly_model),
        ('tree', tree_model),
        ('forest', forest_model),
        ('boosting', boosting_model)
    ],
    final_estimator=RandomForestRegressor(n_estimators=50, random_state=42),
    passthrough=True  # Pass original features along with base model predictions to meta-model
)

# Train base models
linear_model.fit(X_train, y_train)
poly_model.fit(X_train_poly, y_train)
tree_model.fit(X_train, y_train)
forest_model.fit(X_train, y_train)
boosting_model.fit(X_train, y_train)

# Train stacking ensemble model
stacking_ensemble.fit(X_train, y_train)

# Make predictions
y_train_pred_stacking = stacking_ensemble.predict(X_train)
y_test_pred_stacking = stacking_ensemble.predict(X_test)

# Evaluate the stacking model
train_mse_stacking = mean_squared_error(y_train, y_train_pred_stacking)
test_mse_stacking = mean_squared_error(y_test, y_test_pred_stacking)
train_r2_stacking = r2_score(y_train, y_train_pred_stacking)
test_r2_stacking = r2_score(y_test, y_test_pred_stacking)

print("Stacking Ensemble Train Mean Squared Error:", train_mse_stacking)
print("Stacking Ensemble Test Mean Squared Error:", test_mse_stacking)
print("Stacking Ensemble Train R^2 Score:", train_r2_stacking)
print("Stacking Ensemble Test R^2 Score:", test_r2_stacking)
```

```python

# Making predictions for every model
y_train_pred_linear = linear_model.predict(X_train)
y_test_pred_linear = linear_model.predict(X_test)

y_train_pred_poly = poly_model.predict(X_train_poly)
y_test_pred_poly = poly_model.predict(X_test_poly)

y_train_pred_tree = tree_model.predict(X_train)
y_test_pred_tree = tree_model.predict(X_test)

y_train_pred_forest = forest_model.predict(X_train)
y_test_pred_forest = forest_model.predict(X_test)

y_train_pred_boosting = boosting_model.predict(X_train)
y_test_pred_boosting = boosting_model.predict(X_test)

# Plot Function 
def plot_actual_vs_predicted(y_true, y_pred, title):
    plt.figure(figsize=(8, 6))
    plt.scatter(y_true, y_pred, alpha=0.7, color='b')
    plt.plot([y_true.min(), y_true.max()], [y_true.min(), y_true.max()], 'g:') 
    plt.xlabel('Actual Orbital Period') 
    plt.ylabel('Predicted Orbital Period')  
    plt.title(title)
    plt.xlim(y_true.min(), y_true.max())
    plt.ylim(y_true.min(), y_true.max())
    plt.grid(True)
    plt.show()
```

```python
print("For Linear Regression Model: ")
plot_actual_vs_predicted(y_test, y_test_pred_linear, 'Linear Regression: Actual vs Predicted')
```

```python
print("For Polynomial Regression Model: ")
plot_actual_vs_predicted(y_test, y_test_pred_poly, 'Polynomial Regression: Actual vs Predicted')
```

```python
print("For Decision Tree Model: ")
plot_actual_vs_predicted(y_test, y_test_pred_tree, 'Decision Tree: Actual vs Predicted')
```

```python
print("For Random Forest Model: ")
plot_actual_vs_predicted(y_test, y_test_pred_forest, 'Random Forest: Actual vs Predicted')
```

```python
print("For Gradient Boosting Model: ")
plot_actual_vs_predicted(y_test, y_test_pred_boosting, 'Gradient Boosting: Actual vs Predicted')
```

```python
print("For Stacking Ensemble Model: ")
plot_actual_vs_predicted(y_test, y_test_pred_stacking, 'Stacking Ensemble: Actual vs Predicted')
```
